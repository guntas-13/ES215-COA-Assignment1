```python
import matplotlib.pyplot as plt
from latex import latexify
import numpy as np
import pandas as pd

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

latexify(columns = 2)
```

```python
df1 = pd.read_csv("PythonInteger.csv")
df2 = pd.read_csv("PythonFloat.csv")
df3 = pd.read_csv("CPPInteger.csv")
df4 = pd.read_csv("CPPDouble.csv")
T1 = df1.to_numpy().tolist()
T2 = df2.to_numpy().tolist()
T3 = df3.to_numpy().tolist()
T4 = df4.to_numpy().tolist()
mT1, sT1, cT1 = T1[0][1:], T1[2][1:], T1[3][1:]
mT2, sT2, cT2 = T2[0][1:], T2[2][1:], T2[3][1:]
mT3, sT3, cT3 = T3[0][1:], T3[2][1:], T3[3][1:]
mT4, sT4, cT4 = T4[0][1:], T4[2][1:], T4[3][1:]
```

```python
plt.figure(figsize = (10, 6))
plt.plot(mT1, color = "red", marker = "o", label = "Python INTEGER")
plt.plot(mT3, color = "blue", marker = "x", label = "CPP INTEGER")
for i in range(5):
    plt.text(i, mT1[i], f"{mT1[i]}", ha = "center", va = "bottom")
    plt.text(i, mT3[i], f"{mT3[i]}", ha = "center", va = "bottom")
plt.xticks(np.arange(5), np.array([64, 128, 256, 512, 1024]))
plt.xlabel("Matrix Size (N x N)")
plt.ylabel("Time (s)")
plt.title("Time Taken for the MEAT PORTION of Programs for INTEGER Data Type C++ vs Python")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10, 6))
plt.plot(mT2, color = "red", marker = "o", label = "Python FLOAT")
plt.plot(mT4, color = "blue", marker = "x", label = "CPP DOUBLE")
for i in range(5):
    plt.text(i, mT2[i], f"{mT2[i]}", ha = "center", va = "bottom")
    plt.text(i, mT4[i], f"{mT4[i]}", ha = "center", va = "bottom")
plt.xticks(np.arange(5), np.array([64, 128, 256, 512, 1024]))
plt.xlabel("Matrix Size (N x N)")
plt.ylabel("Time (s)")
plt.title("Time Taken for the MEAT PORTION of Programs for FLOAT Data Type C++ vs Python")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10, 6))
plt.plot(cT1, color = "green", marker = "s", label = "Python INTEGER")
plt.plot(cT3, color = "purple", marker = "x", label = "CPP INTEGER")
for i in range(5):
    plt.text(i, cT1[i], f"{cT1[i]}", ha = "center", va = "bottom")
    plt.text(i, cT3[i], f"{cT3[i]}", ha = "center", va = "bottom")
plt.xticks(np.arange(5), np.array([64, 128, 256, 512, 1024]))
plt.xlabel("Matrix Size (N x N)")
plt.ylabel("Time (s)")
plt.title("Total Execution Time of Programs for INTEGER Data Type C++ vs Python")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
